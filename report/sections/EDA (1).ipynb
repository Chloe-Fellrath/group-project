**Association Between Gender and Genre (Chi-Square Test)**

To assess whether gender and musical genre are statistically associated, we conduct a Chi-square 
test of independence. This test evaluates whether the distribution of gender categories differs 
significantly across musical genres. We only include identifiable gender categories and meaningful 
genres, excluding 'unknown' values.


```python
from scipy.stats import chi2_contingency

# -----------------------------------------
# Prepare clean dataset: no unknown gender, no unknown genre
# -----------------------------------------
df_clean = df[
    (df["gender"] != "unknown") &
    (df["genre"]  != "unknown")
]

# -----------------------------------------
# Create contingency table (genre × gender)
# -----------------------------------------
contingency_table = pd.crosstab(df_clean["genre"], df_clean["gender"])
display(contingency_table)

# -----------------------------------------
# Chi-square test
# -----------------------------------------
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)

# Show expected frequencies (optional, informative)
expected_df = pd.DataFrame(expected, 
                           index=contingency_table.index, 
                           columns=contingency_table.columns)

print("\nExpected frequencies:")
display(expected_df.round(2))
```

The Chi-square test evaluates whether gender and musical genre are statistically independent.  
The result shows a Chi-square statistic of 602.65 with 489 degrees of freedom, and a p-value of  
approximately 0.00033, which is far below the conventional 0.05 significance threshold.

This means that we **reject the null hypothesis of independence** and conclude that:

## Gender and musical genre are not independent.  
→ Gender distribution varies significantly across musical genres.

In other words, some genres are disproportionately represented by certain gender categories.  
This statistically confirms the patterns previously observed in the genre-based barplot, where 
rock-oriented genres were heavily male-dominated, while genres such as pop, R&B, and Christmas music 
displayed stronger female or group representation.

This result indicates that gender representation in music is structurally unequal across genres.  
Certain musical styles tend to attract or promote specific gender categories more strongly than others.  
This supports the idea that gender disparities in the music industry are not only temporal (as seen across 
decades) but also genre-specific.

The Chi-square test therefore strengthens the evidence of unequal gender representation and highlights the 
importance of considering genre when analyzing gender-related trends in music datasets.


**Statistical tests for popularity differences across gender categories**

```python
from scipy.stats import levene, f_oneway, kruskal

# Clean data: remove unknown gender & genre
# -----------------------------------------
df_stats = df[
    (df["gender"] != "unknown") &
    (df["genre"]  != "unknown")
]

# Groups for tests
male_pop    = df_stats[df_stats["gender"] == "male"]["track_popularity"]
female_pop  = df_stats[df_stats["gender"] == "female"]["track_popularity"]
group_pop   = df_stats[df_stats["gender"] == "group"]["track_popularity"]
nb_pop      = df_stats[df_stats["gender"] == "non-binary"]["track_popularity"]

# -----------------------------------------
# 1. Levene’s test (equal variances?)
# -----------------------------------------
stat_levene, p_levene = levene(male_pop, female_pop, group_pop, nb_pop)
print("Levene test p-value:", p_levene)

# -----------------------------------------
# 2. If variances equal → ANOVA
# Else → Kruskal–Wallis
# -----------------------------------------
if p_levene >= 0.05:
    print("\nVariances are equal. Running ANOVA...")
    stat_anova, p_anova = f_oneway(male_pop, female_pop, group_pop, nb_pop)
    print("ANOVA p-value:", p_anova)
else:
    print("\nVariances differ. Running Kruskal–Wallis test...")
    stat_kruskal, p_kruskal = kruskal(male_pop, female_pop, group_pop, nb_pop)
    print("Kruskal–Wallis p-value:", p_kruskal)
```

Levene’s test shows a p-value of 0.498, which is far above the 0.05 threshold.  
We therefore fail to reject the null hypothesis of equal variances and proceed with a standard 
one-way ANOVA.

The ANOVA test yields a p-value of 8.06 × 10⁻⁵, which is extremely small and well below the 
0.05 significance threshold. This result indicates that **mean popularity scores differ significantly 
across gender categories**.

In other words, gender is a statistically meaningful factor associated with differences in track 
popularity. This confirms the trends observed in the boxplot, where groups and female artists 
displayed higher median popularity compared to male artists. Non-binary entries remain too sparse 
to form strong conclusions, but they do not affect the overall test outcome.

This finding strengthens the evidence that gender disparities in music are not limited to 
representation (who appears more in the dataset) but also extend to performance outcomes 
(popularity). Gender categories do not share the same popularity distribution, suggesting that 
structural, cultural, or industry-related factors influence which gender groups tend to produce 
more widely consumed or streamed music.
